# Covid US Time Series

```python
import numpy as np
import pandas as pd
```

```python
file_path = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'    
```

```python
US_time_series_data = pd.read_csv(file_path, error_bad_lines=False)
```

```python
US_time_series_data = US_time_series_data.reset_index().set_index('Province_State')
```
